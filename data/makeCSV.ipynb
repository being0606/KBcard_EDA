```python
import os.path
import pandas as pd

# 엑셀 파일 경로
excel_file_path = 'KBdata.xls'

# 엑셀 파일 불러오기
df = pd.read_excel(excel_file_path)

# 5번 인덱스를 열 이름으로 설정 (인덱스 제거)
df.columns = df.iloc[5].values

# 5번 인덱스부터 시작하도록 새로운 데이터프레임 생성
df_new = df[6:].reset_index(drop=True)
df_new
```

```python
# NaN 값만 있는 열들을 삭제
df_deltN = df_new.dropna(axis='columns', how='all')

# 반복되는 열 제외
df_filtered = df_deltN.loc[:, df_deltN.nunique() > 1]
```

```python
# 열 이름 영어로 바꾸기
column_mapping = {
    '이용일': 'Date',
    '이용\n시간': 'Time',
    '이용하신곳': 'Location',
    '국내이용금액\n(원)': 'Domestic Amount (KRW)',
    '해외이용금액\n($)': 'Foreign Amount (USD)',
    '할인금액': 'Discount Amount',
    '적립(예상)\n포인트리': 'Expected Points',
    '상태': 'Status',
    '결제예정일': 'Payment Due Date',
    '승인번호': 'Approval Number'
}
df_final = df_filtered.rename(columns=column_mapping)
df_final
```

```python
csv_file_path = 'dataCard.csv'
df_final.to_csv(csv_file_path, index=False)

print("CSV 파일로 저장되었습니다.")
```
